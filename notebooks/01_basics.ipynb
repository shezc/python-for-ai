# Python 基础语法学习

这个 notebook 演示 Python 的基础语法，适合 Node.js 开发者学习。


## 1. 变量和数据类型


```python
# Python 是动态类型语言，无需声明类型
# Node.js: const name = "Python"
name = "Python"
age = 30
is_learning = True
height = 1.75

print(f"name: {name} (类型: {type(name).__name__})")
print(f"age: {age} (类型: {type(age).__name__})")
print(f"is_learning: {is_learning} (类型: {type(is_learning).__name__})")
print(f"height: {height} (类型: {type(height).__name__})")
```

## 2. 列表（类似 JavaScript 数组）


```python
# Node.js: const fruits = ["apple", "banana", "orange"]
fruits = ["apple", "banana", "orange"]
print(f"fruits: {fruits}")
print(f"第一个元素: {fruits[0]}")
print(f"列表长度: {len(fruits)}")
```

## 3. 字典（类似 JavaScript 对象）


```python
# Node.js: const person = { name: "John", age: 30 }
person = {
    "name": "John",
    "age": 30,
    "city": "Beijing"
}
print(f"person: {person}")
print(f"访问属性: person['name'] = {person['name']}")
print(f"访问属性: person.get('age') = {person.get('age')}")
```

## 4. f-string（类似模板字符串）


```python
# Node.js: `Hello, ${name}!`
name = "Python"
age = 30
message = f"Hello, {name}! You are {age} years old."
print(message)
```

## 5. 列表推导式（类似 map/filter）


```python
# Node.js: [1,2,3,4,5].map(x => x * x).filter(x => x > 10)
numbers = [1, 2, 3, 4, 5]
squared = [x * x for x in numbers]
squared_filtered = [x * x for x in numbers if x * x > 10]

print(f"原始列表: {numbers}")
print(f"平方: {squared}")
print(f"平方且大于10: {squared_filtered}")
```
